```python
#numpy
import numpy as np
# pandas
import pandas as pd
# matplotlib
import matplotlib.pyplot as plt
# the commonly used alias for seaborn is sns
import seaborn as sns

from pandas import DataFrame

# set a seaborn style of my taste
sns.set_style("whitegrid")
# data
df = pd.read_csv( "./data1-a/ab40c054-5031-4376-b52e-9813e776f65e.csv" )
```

```python
# Doing task number 1 here - Removed the rows: '(% Growth over the previous year)' and 
#'GSDP - CURRENT PRICES (` in Crore)' for the year 2016-17.
newdata_df1 = df.drop(df.index[[5,10]])
```

```python
newdata_df2 = newdata_df1.drop(['West Bengal1'], axis = 1)
```

```python
newdata_df2
```

```python
newdata_df3 = newdata_df2.drop([0,1,2,3,4], axis=0)
```

```python
newdata_df3 
```

```python
newdata_df4 = newdata_df3.groupby('Duration').mean()
```

```python
newdata_df5 = newdata_df4.mean()
```

```python
#  the average growth of states over the duration 2013-14, 2014-15 and 2015-16 by taking the mean of the row '(% Growth over previous year)'.
newdata_df4
```

```python
index_ = ['2012-2013', '2013-2014', '2014-2015', '2015-2016']
```

```python
newdata_df5.index = index_
```

```python
newdata_df5
```

```python
growth = newdata_df5.transpose()
```
