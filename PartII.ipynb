```python
# Subsetting only the closing prices column for plotting
curr = df[["Level of Education - State", "Primary - 2014-2015", 'Primary - 2014-2015.1', "Upper Primary - 2014-2015", "Secondary - 2014-2015", "Senior Secondary - 2014-2015"]]
curr.head()
```

```python
cwd = os.getcwd()
dirs = os.listdir( cwd )

df2 = []

for items in dirs:
    if items.find('GSVA')>0 and items.find('.csv')>0:
        x=cwd+'/'+items
        temp_df = pd.read_csv((x), encoding = "ISO-8859-1")
        temp_df = temp_df.loc[::,['S.No.','Item', '2014-15']]
        temp_df['State']=items.split('-')[1]
        df2.append(temp_df)
        
usable_gsva=pd.concat(df2)
```

```python
gsdp = usable_gsva[usable_gsva.Item == 'Per Capita GSDP (Rs.)']
gsdp.head()
```

```python
cor = curr.corr()
round(cor, 3)
```
